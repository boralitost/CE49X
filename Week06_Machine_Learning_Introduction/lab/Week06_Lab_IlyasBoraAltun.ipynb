````python
print(f"Shape before dropping NA: {df.shape}")
print("Missing per column:\n", df.isnull().sum())
df_clean = df.dropna().copy()
print(f"Shape after dropping NA:  {df_clean.shape}")
df = df_clean
````

````python
X = df.drop('target', axis=1)
y = df['target']

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42
)

print("X_train:", X_train.shape, " X_test:", X_test.shape)
print("y_train:", y_train.shape, " y_test:", y_test.shape)
````

````python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

print("Scaled train shape:", X_train_scaled.shape)
print("Scaled test shape: ", X_test_scaled.shape)
````

### Data Leakage Explanation

We **fit** `StandardScaler` on `X_train` only because the scaler learns the mean and standard deviation from data the model is allowed to see during training. The test set must stay **unseen** until evaluation. If we fit the scaler on the full dataset before splitting, statistics from test patients leak into preprocessing. That inflates test performance and is **data leakage**: the pipeline indirectly uses test information, so reported accuracy is optimistically biased.


````python
lr = LogisticRegression(max_iter=1000, random_state=42)
t0 = time.time()
lr.fit(X_train_scaled, y_train)
lr_time = time.time() - t0

lr_train_acc = lr.score(X_train_scaled, y_train)
lr_test_acc = lr.score(X_test_scaled, y_test)

print("Logistic Regression")
print(f"  Training accuracy: {lr_train_acc:.4f}")
print(f"  Test accuracy:     {lr_test_acc:.4f}")
print(f"  Training time:     {lr_time:.4f} s")
````

````python
dt = DecisionTreeClassifier(random_state=42)
t0 = time.time()
dt.fit(X_train_scaled, y_train)
dt_time = time.time() - t0

dt_train_acc = dt.score(X_train_scaled, y_train)
dt_test_acc = dt.score(X_test_scaled, y_test)

print("Decision Tree")
print(f"  Training accuracy: {dt_train_acc:.4f}")
print(f"  Test accuracy:     {dt_test_acc:.4f}")
print(f"  Training time:     {dt_time:.4f} s")
````

### Model Comparison

On the **test set**, logistic regression often generalizes slightly better than an unrestricted decision tree on this dataset. The tree may reach **very high training accuracy** with a noticeably larger train–test gap.

**Overfitting:** If training accuracy is much higher than test accuracy (especially for the tree), the model has **overfit** the training noise. The **gap** between train and test accuracy is the main signal: a small gap suggests good generalization; a large gap suggests overfitting. Logistic regression is more constrained and typically shows a **smaller gap** than the default decision tree here.

### Confusion Matrix Interpretation

Sklearn uses rows = **actual** class, columns = **predicted** class (both ordered 0 then 1):

- **True Negatives (TN):** `cm[0,0]` — model said no disease and the patient truly had no disease.
- **False Positives (FP):** `cm[0,1]` — model predicted disease but the patient was healthy (unnecessary follow-up / anxiety).
- **False Negatives (FN):** `cm[1,0]` — model said no disease but the patient had disease (**risky miss**).
- **True Positives (TP):** `cm[1,1]` — model correctly flagged heart disease.

Run the code cell above and read the printed line `TN=..., FP=..., FN=..., TP=...` for the exact counts on your split.


````python
print(classification_report(y_test, y_pred, target_names=['No disease (0)', 'Heart disease (1)']))
````

### Classification Report Interpretation

- **Precision for heart disease (class 1):** Of everyone the model labels as diseased, what fraction truly has disease. High precision means fewer **false alarms** (FP).
- **Recall for heart disease (class 1):** Of everyone who truly has disease, what fraction the model finds. High recall means fewer **missed cases** (FN).
- **Which matters more here:** The director prefers missing as few sick patients as possible, so **recall** is usually more important than precision for initial screening, even if that means more follow-up tests.


````python
cv_scores = cross_val_score(best_model, X_train_scaled, y_train, cv=5, scoring='f1')
print("5-fold F1 scores:", cv_scores)
print(f"Mean F1: {cv_scores.mean():.4f} (+/- {cv_scores.std():.4f})")
````

### Cross-Validation Interpretation

The **mean 5-fold F1** on training folds should be in a similar ballpark to what you see on a single held-out test split (not identical). If the **standard deviation across folds is large**, estimates are unstable—the model’s performance depends a lot on which patients ended up in each fold, so we should not over-trust one lucky random split.


````python
# Part D: Overfitting curve - Decision Tree with varying max_depth

depths = [1, 2, 3, 5, 8, 12, 20, None]
train_accs = []
test_accs = []

for d in depths:
    dt_d = DecisionTreeClassifier(max_depth=d, random_state=42)
    dt_d.fit(X_train_scaled, y_train)
    train_accs.append(dt_d.score(X_train_scaled, y_train))
    test_accs.append(dt_d.score(X_test_scaled, y_test))

depth_labels = [str(d) if d is not None else 'None' for d in depths]
for lab, tr, te in zip(depth_labels, train_accs, test_accs):
    print(f"max_depth={lab:>4}  train_acc={tr:.4f}  test_acc={te:.4f}")
````

````python
# Plot the overfitting curve: train and test accuracy vs max_depth
depth_labels = [str(d) if d is not None else 'None' for d in depths]
fig, ax = plt.subplots(figsize=(9, 5))
x = np.arange(len(depths))
ax.plot(x, train_accs, 'o-', label='Training accuracy', color='#DC2626')
ax.plot(x, test_accs, 's-', label='Test accuracy', color='#2563EB')
ax.set_xticks(x)
ax.set_xticklabels(depth_labels)
ax.set_xlabel('max_depth')
ax.set_ylabel('Accuracy')
ax.set_title('Decision tree: training vs test accuracy vs tree depth')
ax.legend(frameon=False)
ax.grid(True, linestyle='--', alpha=0.35)
plt.tight_layout()
plt.show()
````

### Overfitting Curve Interpretation

Very small **max_depth** tends to **underfit**: both training and test accuracy stay limited. As depth grows, training accuracy often climbs toward 1.0 while test accuracy rises to a peak and may then fall—the widening train–test gap is **overfitting** (memorizing noise), as in the lecture.

Choose a **max_depth** near the best **test** accuracy on the plot; in practice you would confirm with cross-validation.
